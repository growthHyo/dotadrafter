```python
from peewee import *
import datetime
import requests

api_key = 'FE70CE9FC0D6D99279498CE852587F59'

db = SqliteDatabase('data/doto.db')

r = requests.get('https://api.steampowered.com/IEconDOTA2_570/GetHeroes/V001/?key=' + api_key)

class Hero(Model):
    id = IntegerField(primary_key=True)
    name = CharField()
    class Meta:
        database = db
        
class Match(Model):
    id = IntegerField(primary_key=True)
    date = DateTimeField(default=datetime.datetime.now)
    seq_num = IntegerField()
    radiant_heroes = CharField()
    dire_heroes = CharField()
    radiant_win = BooleanField()
    class Meta:
        database = db
        
db.connect()
Hero.drop_table(True)
Hero.create_table()
        
for hero in r.json()['result']['heroes']:
    h = Hero(id=hero['id'], name=hero['name'][14:])
    h.save(force_insert=True)
        

for h in Hero.select():
    print(h.name)
    print(h.id)
    
print(len(Hero.select()))

db.close()
```
